初級 Lv3：input() と型変換（超重要）

次の条件を満たすコードを書け。

条件

ユーザーに Age? （最後に半角スペースあり）と聞く

入力された年齢を 整数に変換する

次の形式で表示する（完全一致）
Next year: <age+1>
例：入力が 20 なら Next year: 21

ヒント（考え方の順序）

input() の返り値は 文字列 だ。油断するな

int() で変換してから + 1

例の形：

age_str = input(...)

age = int(age_str)

print(...)

````python
age_str = input("Age? ")
age_int = int(age_str)

print("Next year:",  age_int + 1)
````

条件

ユーザーに Name? と聞いて名前を受け取る

ユーザーに Age? と聞いて年齢を受け取り、整数に変換

f文字列を使って、次の1行を表示する（形式厳守）
Hi <name>, you will be <age+1> next year.

ヒント

f"Hi {name}, you will be {age + 1} next year."

{} の中は式を書ける（age + 1 とかな）

````python
user_name = input("Name? ")
user_age_str = input("Age? ")
user_age = int(user_age_str)

print(f"Hi {user_name}, you will be {user_age + 1} next year.")
````

初級 Lv6：改行なし出力（printのend）

次の条件を満たすコードを書け。

条件

初級 Lv8：ミニ対話＋計算（まとめ） 行くぞ。

課題

Width? と聞いて幅を入力させる（整数に変換）

Height? と聞いて高さを入力させる（整数に変換）

長方形の面積 Area: <width*height> を表示する

ヒント

input() → int() → 計算

出力は Area: 42 みたいな形（コロン後は半角スペース1つ）

````python
width_str = input("Width? ")
height_str = input("Height? ")

width = int(width_str)
height = int(height_str)

area = width * height
print("Area:",area)
````

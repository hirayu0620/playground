中級 Lv10：総合（FizzBuzzの入口）

条件

3の倍数：Fizz

5の倍数：Buzz

3と5の両方の倍数：FizzBuzz

ヒント（超重要）

上級 Lv2：デフォルト引数（関数設計の基本）

次の条件を満たすコードを書け。

条件

関数 greet(name, prefix="Hello") を作る

戻り値として文字列を返す："<prefix>, <name>!"

ユーザーに Name? と聞いて name を受け取る

ユーザーに Prefix? と聞いて prefix を受け取る（空文字ならデフォルトを使う）

最後に関数の戻り値を print() で表示する

ヒント（重要）

prefix_input = input("Prefix? ")

if prefix_input == "": のときは greet(name)

それ以外は greet(name, prefix_input)

````python
def greet(name, prefix="Hello"):
    return f"{prefix}, {name}!"

name = input("Name? ")
prefix = input("Prefix? ")

if prefix == "":
    result = greet(name)
else: 
    result = greet(name, prefix)

print(result)
````
